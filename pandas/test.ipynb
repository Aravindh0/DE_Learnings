```python
#concat

df1 = pd.DataFrame({'id' : [1,2,3,4,5,6],
                   'name' : ['a', 'b', 'c', 'd', 'e', 'f']})
df2 = pd.DataFrame({'id' : [7,8,9,10],
                   'name' : ['g', 'h', 'i', 'j']})
df_all = [df1,df2]
df_all
```

```python
pd.concat([df,df1]).tail(25)
```

```python
result = pd.concat(df_all, axis = 0) #axis = 0 refers horizontal or row
display(result)
```

```python
result = pd.concat(df_all, axis = 1) #axis = 1 refers to vertical or columns
display(result)

#The decimals in the id columns are due to the presence of NaN values introduced by the differing row counts in df1 and df2. Pandas converts integers to floats to accommodate NaN values.
```

```python
pd.concat([df1,df2], keys = ['a', 'b']) #the keys acts as a hierarchical index
```

```python
df3 = pd.DataFrame({'id' : [1,2,3,4,7],
                   'name' : ['a', 'i', 'c', 'd', 'm']})
```

```python
#merge (joins)

pd.merge(df1,df3, how = 'left', on = 'id')
```

```python
pd.merge(df1, df3, how = 'right', on = 'id')
```

```python
pd.merge(df1, df3, how = 'inner', on = 'id')
```

```python
pd.merge(df1, df3, how = 'outer', on = 'id')
```

```python
df4 = pd.DataFrame ({'s_id' : [1,2,3,8,9],
                     'name' : ['u', 'g', 'k', 'n', 'v']})
```
